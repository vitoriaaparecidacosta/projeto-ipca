```python
import requests
import pandas as pd

# Cabeçalho
serie = 1619  # Salário
url = f"https://api.bcb.gov.br/dados/serie/bcdata.sgs.{serie}/dados?formato=json"

# Requisição
dados = requests.get(url)
df = pd.DataFrame(dados.json())

# Convertendo em dataframe spark
df_spark = spark.createDataFrame(df)
tabela_destino = "datalakehouse.bronze.bcb_salario_serie_1619"  

# Criando tabela SQL
df_spark.write.format("delta") \
    .mode("overwrite") \
    .saveAsTable(tabela_destino)
```
