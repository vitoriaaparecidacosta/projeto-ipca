```python
%sql
-- Cria a tabela calend√°rio
CREATE OR REPLACE TABLE datalakehouse.bronze.calendario AS
WITH dates AS (
    SELECT sequence(to_date('1980-01-01'), to_date('2025-12-31'), interval 1 day) AS date_seq
),
exploded AS (
    SELECT explode(date_seq) AS date
    FROM dates
)
SELECT 
    date,
    year(date) AS ano,
    month(date) AS mes,
    day(date) AS dia,
    quarter(date) AS semestre,
    dayofweek(date) AS num_dia_semana,
    date_format(date, 'E') AS nome_dia_semana,
    weekofyear(date) AS semana_ano
FROM exploded;
```
